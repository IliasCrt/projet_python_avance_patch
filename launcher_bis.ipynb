```python
import pygame as pg
import random as rd
import time
import math
from game import Game

pg.init()


# Création de la classe Game
game=Game()

# Initialisation de la fenêtre du jeu
#screen = pg.display.set_mode((1200, 500))
screen = pg.display.set_mode((1080, 720))
pg.display.set_caption("Affichage dans le coin")

#Téléchargement de l'image de l'icone du jeu et la convertir 
icone = pg.image.load("smiley.jpg").convert() 

#Nommer la fenêtre et placer l'icone
pg.display.set_caption("GeoMinetry Dash") 
pg.display.set_icon(icone)


#Interface graphique
accueil = pg.image.load("Accueil.png")

#Boutons de commande
play_button = pg.image.load("bouton.png")
play_button = pg.transform.scale(play_button,(300,300))
play_button_rect = play_button.get_rect()
play_button_rect.x= 390
play_button_rect.y= 200

# Variables pour la gestion du temps
temps_actuel = pg.time.get_ticks()
temps_suivant = temps_actuel + rd.randint(1000, 5000)  # Intervalle initial aléatoire en millisecondes
first = 0
# Définition de variables et initialisation de Pygame
clock = pg.time.Clock()
running = True

# Boucle principale du jeu
while running :
    t=time.time()
    clock.tick(100) # Limitation de la vitesse de la boucle principale à 100 images/s
    #Vérifier si le jeu a commencé ou non
    if game.is_playing :
        game.update(t0,t,screen)
        
    else : 
        screen.blit(accueil, (0,0))
        screen.blit(play_button, (390, 200))
        police_1 = pg.font.Font(None, 100)
        a=int(max(game.L)/100)
        texte_score = "Meilleur score : {}".format(str(a))
        texte_surface_score = police_1.render(texte_score, True, (0, 0, 0))
        coin_x1, coin_y1 = 250,100 
        screen.blit(texte_surface_score, (coin_x1, coin_y1))
        for event in pg.event.get():
            if event.type == pg.QUIT:
               running = False
            elif event.type == pg.KEYDOWN:
                #Si la touche est "Q" on veut quitter le programme
                if event.key == pg.K_q:
                    running = False   
            if event.type == pg.MOUSEBUTTONDOWN :
            # Si la souris clique sur le bouton, le jeu démarre
                if play_button_rect.collidepoint(event.pos) :
                    t0 = time.time()
                    game.is_playing = True

        
    pg.display.flip() # Rafraîchissement de l'écran
    
pg.quit() # Fermeture de Pygame à la fin du jeu
```
